```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/sivabalanb/Data-Analysis-with-Pandas-and-Python/master/nba.csv')
```

```python
df.head()
```

#Data Preparation

```python
df.info()
```

```python
df.Age.unique()
```

```python
df.Age = df.Age.fillna(df.Age.mean().round(0))
```

```python
df.Weight.unique()
```

```python
df.Weight = df.Weight.fillna(df.Weight.mean().round(0))
```

```python
df.Number = df.Number.fillna(df.Number.mean().round(0))
```

```python
df.Number.unique()
```

```python
df.Position.mode()
```

```python
df.Team.unique()
```

```python
df.Team = df.Team.fillna(df.Team.mode()[0])
```

```python
df.Height.unique()
```

```python
df.Height.mode()
```

```python
df.Height = df.Height.fillna(df.Height.mode()[0])
```

```python
df.Salary = df.Salary.fillna(df.Salary.mean().round(0))
```

#Boxplot

```python
import matplotlib.pyplot as plt

#Select variable from dataframe
data = df.Age

#Creating plot
plt.boxplot(data)
plt.xlabel("Age")
plt.title("Boxplot of Age")

#Show plot
plt.show()
```

```python
#Select several numerical variables from dataframe
data = [df.Age, df.Number]

fig = plt.figure(figsize = (10,7))

#creating axes instance
ax = fig.add_axes([0,0,1,1])

#creating plot
bp = ax.boxplot(data)

#show plot
plt.show()
```

#Histogram

```python
#Select variable from the dataset
data = df.Age

#Creating plot
plt.hist(data, bins = 8)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Age")

#Show plot
plt.show()
```

```python
#Multiple Histogram
#Digunakan untuk melihat distribusi data diantara dua entitas variable

plt.hist(df.Age,
         facecolor = 'orangered',
         edgecolor = 'maroon',
         bins = 10)

plt.hist(df.Number,
        facecolor = 'peru',
        edgecolor = 'blue', #kerangka garis histogram
        bins = 15,
        alpha = 0.5)

plt.title("Histogram of Age and Number")

#Adjust the spacing bertween subplots
plt.tight_layout()

#show the plot
plt.show()
```

#Bar plot

```python
#Extract the "Position" column from the dataset
position_data = df['Position']

#Get unique values
unique_positions = position_data.unique()

#Count the value of each Position
position_counts = position_data.value_counts().sort_values(ascending=True)

#Define bar labels
bar_labels = [f"Position (position)" for position in unique_positions]

#Creat a bar chart
plt.bar(bar_labels, position_counts)

#Add title and axis labels
plt.title("Bar Plot of Position")
plt.xlabel("Position")
plt.ylabel("Count")

#Rotate x-axis labels for better readability
plt.xtricks(rotation=45)

#Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Extract the "Position" column from the dataset
position_data = df['Position']

# Get unique values of Position and drop NaN values
unique_positions = position_data.dropna().unique()

# Count the value of each Position, dropping NaN values
position_counts = position_data.value_counts().sort_values(ascending=True)

# Filter position_counts to only include valid positions
position_counts = position_counts[unique_positions]

# Create a bar chart
plt.bar(position_counts.index, position_counts.values, color = '#33d4ff')

# Add title and axis labels
plt.title("Bar Plot of Position")
plt.xlabel("Position")
plt.ylabel("Count")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
#Extract the "Position" and "Wight"columns from the dataset
position_data = df['Position']
weight_data = df['Weight']

#Group the data by Position and calculate the averange weight
average_weight_by_position = weight_data.groupby(position_data).mean().sort_values(ascending=True)

#Define bar labels
bar_labels = average_weight_by_position.index.to_list()

#creat a bar chart
plt.bar(bar_labels, average_weight_by_position)

#add title and axis labels
plt.title("Bar Plot of Position vs. Weight")
plt.xlabel("Position")
plt.ylabel("Average Weight")

#Rotate x-axis labels for better readability
plt.xticks(rotation=45)

#Show the plot
plt.show()
```

#Pie chart

```python
#Extract the "Position" column from the dataset
position_data = df['Position']

#Get unique values of Position
unique_positions = position_data.unique()

#Count the frequency of each Position
position_counts = position_data.value_counts()

#Define labels for the pie chart
labels = [f"Position (position)" for position in unique_positions]

#Create a pie chart
plt.pie(position_counts, labels=labels, autopct='%1.1f%%')

#Add title
plt.title("Pie Chart of Position")

#Show the plot
plt.show()
```

```python
#Define the start and end_dates
start_date = "2020-01-01"
end_date = "2024-06-14"

#Download Apple stock data
apple_data = yf.download("AAPL", start=start_date, end=end_date)

#Calculate monthly average closing prices
monthly_average_prices = apple_data['Close'].resample('BM').mean()

#Create a line chart
plt.plot(monthly_average_prices)

#Add labels and title
plt.ylabel("Closing Price")
plt.xlabel("Date")
plt.title("Appple Stock Price (2020-01-01 to 2024-06-14)")

#Rotate x-axis labels for better readability
plt.xticks(rotation=45)

#Show the plot
plt.show()
```
